```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
img = plt.imread('Image of fingerprint.png')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
new_img = img.copy()
new_img = new_img/255.0
for i in range(img.shape[0]):
    for j in range (img.shape[1]):
        for k in range (0,3):
            if new_img[i][j][k]>0.0015:
                new_img[i][j][k]=1
            else:
                new_img[i][j][k]=0
```

```python
plt.imshow(new_img)
```

#### Taking a Kernal of 3x3 size

```python
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])
plt.imshow(kernel, cmap='gray')
print(kernel.shape)
plt.show()
```

```python
n,m = new_img.shape[0],new_img.shape[1]
erosion_img = new_img.copy()
for i in range (1,n-1):
    for j in range (1,m-1):
        temp = new_img[i-1:i+2, j-1:j+2]
        product = temp*kernel
        erosion_img[i,j]=np.min(product)
```

```python
plt.imshow(erosion_img)
plt.show()
```

```python
n,m = new_img.shape[0],new_img.shape[1]
dilation_img = new_img.copy()
for i in range (1,n-1):
    for j in range (1,m-1):
        temp = new_img[i-1:i+2, j-1:j+2]
        product = temp*kernel
        dilation_img[i,j]=np.max(product)
```

```python
plt.imshow(dilation_img)
plt.show()
```

```python
n,m = new_img.shape[0],new_img.shape[1]
opening_img = new_img.copy()
for i in range (1,n-1):
    for j in range (1,m-1):
        temp = erosion_img[i-1:i+2, j-1:j+2]
        product = temp*kernel
        opening_img[i,j]=np.max(product)
        
#Tahmim_Hossain_1610028
```

```python
plt.imshow(opening_img)
plt.show()
```

```python
n,m = new_img.shape[0],new_img.shape[1]
closing_img = new_img.copy()
for i in range (1,n-1):
    for j in range (1,m-1):
        temp = dilation_img[i-1:i+2,j-1:j+2]
        product = temp*kernel
        closing_img[i,j]=np.min(product)
        
#Tahmim_Hossain_1610028
```

```python
plt.imshow(closing_img)
plt.show()
```
