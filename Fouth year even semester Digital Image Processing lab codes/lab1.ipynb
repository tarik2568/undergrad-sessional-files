```python
import matplotlib.pyplot as plt
```

```python
rgb = plt.imread('fg.jpg')
```

```python
plt.imshow(rgb)
rgb.shape
```

```python
r = rgb[:,:,0]
g = rgb[:,:,1]
b = rgb[:,:,2]
```

```python
import numpy as np
def histg(img):
    freq = np.zeros(256,dtype = int)
    for i in range(0,img.shape[0]):
        for j in range(0,img.shape[1]):
            freq[img[i][j]] = freq[img[i][j]] + 1
            
            
    intencity = np.zeros(256,dtype=int)
    
    for k in range(0,256):
        intencity[k] = k
        
    plt.plot(intencity,freq)
        
```

```python
histg(r)
```

```python
histg(g)
```

```python
histg(b)
```

```python
rc = r.copy()
```

```python
for i in range(0,r.shape[0]):
        for j in range(0,r.shape[1]):
            if r[i][j]>155 :
                rc[i][j] = 255
            else:
                rc[i][j] = 0
```

```python
plt.imshow(rc,cmap = 'gray')
```

```python
gc = g.copy()
```

```python
for i in range(0,g.shape[0]):
        for j in range(0,g.shape[1]):
            if g[i][j]>125 :
                gc[i][j] = 255
            else:
                gc[i][j] = 0
```

```python
plt.imshow(gc,cmap = 'gray')
```

```python
bc = b.copy()
```

```python
for i in range(0,b.shape[0]):
        for j in range(0,b.shape[1]):
            if b[i][j]>80 :
                bc[i][j] = 255
            else:
                bc[i][j] = 0
```

```python
plt.imshow(bc,cmap = 'gray')
```

```python
negg = 255 - gc
```

```python
plt.imshow(negg,cmap = 'gray')
```

```python
newpic = np.zeros((r.shape[0],r.shape[1]),dtype = int)
```

```python
for i in range(0,b.shape[0]):
        for j in range(0,b.shape[1]):
            if rc[i][j] == 255 :
                if negg[i][j] == 255 :
                    if bc[i][j] == 255 :
                        newpic[i][j] = 255
            else:
                newpic[i][j] = 0
```

```python
plt.imshow(newpic,cmap = 'gray')
```

```python
mainpic = rgb.copy()
```

```python
for i in range(0,b.shape[0]):
        for j in range(0,b.shape[1]):
            if newpic[i][j] == 255 :
                mainpic[i][j][0]=255
                mainpic[i][j][1]=255
                mainpic[i][j][2]=255
```

```python
plt.imshow(mainpic,cmap = 'gray')
```
